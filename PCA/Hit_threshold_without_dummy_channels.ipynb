```python
def module_names():
    names = []
    data = pd.read_csv(data_filesT[0] ,sep=' ', header=None, engine='python', error_bad_lines=False, warn_bad_lines=False)
    data = data[data[0].str.contains("low_threshold")]
    names = data[1]
    return names
```

Tutaj przekleiłem potrzebne funkcje do redukcji dummy channels z pliku calibration_dataset.py.

```python
def find_dummy_channels(dataset : pd.DataFrame):
    blocked = (dataset==127.0)
    trully_dummy = []
    for channel in blocked:
        column = blocked[channel]
        if column.all():
            trully_dummy.append(channel)
    return trully_dummy

def remove_dummy_channels(data : pd.DataFrame):
    dummy_channels = find_dummy_channels(data)
    channels = data.drop(dummy_channels, axis=1)
    #assert channels.iloc[:,9:].shape[1] == 2048, "Number of channelsis not equal to 2048, found {}".format(channels.iloc[:,9:].shape[1])
    #columns = ["channel{}".format(i) for i in range(channels.iloc[:,9:].shape[1])]
    #channels.columns = channels.columns[:9].tolist()+columns
    return channels
```

Każdy wyseparowany zestaw danych zawiera 84x2304 danych, w tym przypadku wartości hit treshold.

```python
for file in data_filesT:
    data2 = pd.DataFrame()
    data = pd.read_csv(file ,sep=' ', header=None, engine='python', error_bad_lines=False, warn_bad_lines=False)
    data = data[data[0].str.contains("hit_threshold")]
    data2 = data.loc[:, 2:2305]

    data2.index = module_names()
    data2.columns = range(2304)

    print("{0}".format(str(datetime.datetime.strptime(os.path.basename(file), '%Y_%m_%d-%H_%M_%S.csv').strftime("%Y/%m/%d, %H:%M:%S"))))
    print("Data shape before dummey channels removal:")
    print(data2.shape)

    data3 = remove_dummy_channels(data2)
    print("Data shape after dummey channels removal:")
    print(data3.shape)
    
    print("")
    print("__________________________________________")
    print("")

```
